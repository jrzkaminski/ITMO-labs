```python
def measure_time(n_experiments, method, g, source, dest):
    times = []
    for i in range(10):
        start_time = timer()
        method(g, source, dest, 'weight')
        times.append(timer() - start_time)
    
    return np.array(times, dtype=np.float64).mean(axis=0)


print('Dijkstra time {:.6f}'.format(measure_time(10, nx.algorithms.shortest_paths.weighted.dijkstra_path, g, 0, 99)))
print('Bellman-Ford time {:.6f}'.format(measure_time(10, nx.algorithms.shortest_paths.weighted.bellman_ford_path, g, 0, 99)))
```
